```python
#Import Necessary Libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler , LabelEncoder

#Read DataSet
df = pd.read_csv("dataset.csv")

# #Encode Catagorical features
# catcol = ['Income_type', 'Education_type', 'Family_status', 'Housing_type','Occupation_type']
# for col in catcol:
#   le = LabelEncoder()
#   df[col] =le.fit_transform(df[col])

# #Featur scaling
# sc = StandardScaler()
# NumCol = ['Total_income','Age','Years_employed', 'Num_children','Num_family','Account_length']
# df[NumCol] = sc.fit_transform(df[NumCol])

# df=df.drop(columns=['ID'])

#Plotting Histogram
import matplotlib.pyplot as plt

def histo_plot():
  df.dropna(inplace = True)
  plt.figure(figsize=(10,6))

  plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
  plt.title('Age Distribution')
  plt.xlabel('Age')
  plt.ylabel('Frequency')
  plt.grid(True)
  plt.show()

histo_plot()

#Barchart
def barch():
  plt.figure(figsize=(10,6))
  df['Target'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
  plt.title('Count Of Target Variable')
  plt.xlabel('Target')
  plt.ylabel('Count')
  plt.xticks(ticks=[0,1], labels=['No', 'Yes'])
  plt.grid(axis='y')
  plt.show()
barch()

#Scatter Plot
def scatterIncAge():
  plt.figure(figsize=(10,6))
  plt.scatter(df['Age'],df['Total_income'], color = 'skyblue')
  plt.title('Total income vs. Age')
  plt.xlabel('Age')
  plt.ylabel('Total Income')
  plt.grid(True)
  plt.show()
scatterIncAge()


#Convert total income to numeric if it is not

df['Total_income'] = pd.to_numeric(df['Total_income'], errors='coerce')
#Box plot
def boxppl():
  plt.figure(figsize=(10,6))
  df.boxplot(column='Total_income', by = 'Education_type', grid=False)
  plt.title('Total income by education type')
  plt.xlabel('Education Type')
  plt.suptitle('')
  plt.ylabel('Total Income')
  plt.xticks(rotation=45)
  plt.show()
boxppl()

#Heatmap of correlation Matrix
#Encode Catagorical features
catcol = ['Income_type', 'Education_type', 'Family_status', 'Housing_type','Occupation_type']
for col in catcol:
  le = LabelEncoder()
  df[col] =le.fit_transform(df[col])

#Featur scaling
sc = StandardScaler()
NumCol = ['Total_income','Age','Years_employed', 'Num_children','Num_family','Account_length']
df[NumCol] = sc.fit_transform(df[NumCol])

df=df.drop(columns=['ID'])

import seaborn as sns
def plotheatmap():
  plt.figure(figsize=(10,6))
  correlation_matrix = df.corr()
  sns.heatmap(correlation_matrix, annot = True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
  plt.title('Correlation Matrix Heatmap')
  plt.show()
plotheatmap()

#ploting Elbow plot for k mean clutering
from sklearn.cluster import KMeans
import numpy as np

def plotelbow():
  sse = []
  k_range = range(1,11)
  for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df.drop(columns=['Target']))
    sse.append(kmeans.inertia_)
  plt.figure(figsize=(10,6))
  plt.plot(k_range,sse,marker='o')
  plt.title('Elbow Plot for K-Mean clustering')
  plt.xlabel('Nummber Of clusters')
  plt.ylabel('Sum of squared errors')
  plt.grid(True)
  plt.show()

plotelbow()

#Line Fitting
from sklearn.linear_model import LinearRegression
def plot_linefitt():
  x=df[['Age']].values
  y = df['Total_income'].values

  #Fit a linear Model
  lin_reg = LinearRegression()
  lin_reg.fit(x, y)

  #Predictions
  y_pred = lin_reg.predict(x)

  plt.figure(figsize = (10,6))
  plt.scatter(x, y , color='skyblue', label = 'Data Points')
  plt.plot(x,y_pred, color='red', linewidth=2,label ='Linear fit')
  plt.title('Linear Fit for total income vs. age')
  plt.xlabel('Age')
  plt.ylabel('Total Income')
  plt.legend()
  plt.grid(True)
  plt.show()

plot_linefitt()

```
